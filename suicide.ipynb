```python
from zipfile import ZipFile
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_name = 'suicide-rates-overview-1985-to-2016.zip'

#Open zip file in read mode
with ZipFile(file_name,'r') as zip:
    #print the contents of the zip file
    zip.printdir()
        
    #extract all the files
    zip.extractall()
    print('\n.csv file extracted')
```

```python
#Read .csv file
data = pd.read_csv('master.csv')
data.head()
```

```python
# Sum the suicide_no
total_suicide = data.suicides_no.sum()

# Group the Data Frame by 'sex' and sum the suicides no
total_by_sex = data.groupby(['sex']).agg({'suicides_no' : 'sum'})

# Find the percentage ratio between the male and female
percentage_by_sex = total_by_sex.apply(lambda x:
                                      100* x / float(x.sum()))
```

```python
# Pie chart with pandas
percentage_by_sex.plot.pie(figsize = (5,5),
                           subplots= ("Male", "Female"),
                           colors = ("r", "b"),
                           title = 'Percentage number of suicide by sex',
                           explode = (0.15, 0),
                           startangle = 7,
                           autopct = '%2.1f%%')
```

```python
# Describe the percentage of male and female
male = percentage_by_sex.stack().male
female = percentage_by_sex.stack().female

proportions = [male, female]

# Create figure with custom size
plt.figure(figsize = (5,5))

# Create a pie chart with matplotlib
plt.pie(
    proportions, 
    labels = ['Male', 'Female'],
    colors = ['b', 'r'],
    explode = (0.15, 0),
    startangle = 90,
    autopct = '%2.1f%%')

# Set the title, legend 
plt.title('Percentage number of suicide by sex', fontsize = 12)

plt.legend( title="Ingredients",
            loc="center left",
            bbox_to_anchor=(1, 0, 0.5, 1),
            )

# View
plt.tight_layout()
plt.show()
```

```python
# Define suicides no of the male and female by age
male_age = data.groupby(['sex', 'age']).agg({'suicides_no' : 'sum'}).drop(labels = 'female')
female_age = data.groupby(['sex', 'age']).agg({'suicides_no' : 'sum'}).drop(labels = 'male')

# Rename for better sort
male_age = male_age.rename({'5-14 years':'05-14 years'}).sort_index()
female_age = female_age.rename({'5-14 years':'05-14 years'}).sort_index()

# Prepare DataFrame for plot 
male_age = male_age.reset_index(level = ['sex'], drop = True).rename(columns={ 'suicides_no': 'Male'})

female_age = female_age.reset_index(level = ['sex'], drop = True).rename(columns={ 'suicides_no': 'Female'})
```

```python
# Delete index name of age for not display in the bar
del male_age.index.name
del female_age.index.name

# Bar plot with pandas 
pd.concat([male_age, female_age], axis = 1).plot.bar(color = ['b', 'r'], 
                                                     width = 0.50, 
                                                     title = 'Suicides no of the male and female by age', 
                                                     rot = 45)
```

```python
# Create figure 
fig, ax = plt.subplots()

# Set the bar chart width and index
width = 0.25
ind = np.arange(len(male_age.index))

# Bar chart with matplotlib
female_bar = ax.bar(ind, 
                    female_age.Female, 
                    color ='r', 
                    width = width)

male_bar = ax.bar(ind - width, 
                  male_age.Male, 
                  color ='b', 
                  width = width)

# Set the title, xticks, xticklabels, legend
ax.set_title('Suicides no of the male and female by age')
ax.set_xticks(ind - width / 2)
ax.set_xticklabels(male_age.index, rotation = 45)
ax.legend((male_bar[0], female_bar[1]), ('Male', 'Female'))
```

```python
# Define suicides no by the male and female in Turkey
tr_male = data.groupby(['country','sex']).get_group(('Turkey','male')).groupby(['year']).agg({'suicides_no': 'sum'})
tr_female = data.groupby(['country','sex']).get_group(('Turkey','female')).groupby(['year']).agg({'suicides_no': 'sum'})

# Rename columns
tr_male = tr_male.rename(columns={ 'suicides_no': 'Male'})
tr_female = tr_female.rename(columns={ 'suicides_no': 'Male'})
```

```python
# Plot with pandas
pd.concat([tr_male, tr_female], axis = 1).plot(color = ['b', 'r'], 
                                                   title = 'Suicides in Turkey', 
                                                   marker = 'o' )
```

```python
# Create figure with custom size
fig, ax = plt.subplots()

# Plot with mathplotlib
male_plt = plt.plot(tr_male, 
                    marker = 'o', 
                    color='b')

female_plt = plt.plot(tr_female, 
                      marker ='o', 
                      color='r')

# Set the title,legend
ax.set_title('Suicides in Turkey')
ax.legend((male_plt[0], female_plt[0]), ('Male', 'Female'))

# View
plt.show()
```
